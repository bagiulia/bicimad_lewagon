```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("../raw_data/weather_madrid.csv")
df.head(3)
```

```python
df['hour'] = df.apply(lambda x:datetime.strptime(x['dt_iso'], '%Y-%m-%d %H:%M:%S +0000 UTC').hour, axis=1)
df['day'] = df.apply(lambda x:datetime.strptime(x['dt_iso'], '%Y-%m-%d %H:%M:%S +0000 UTC').date(), axis=1)
```

```python
df["datetime"] = df["day"].astype(str)+df["hour"].astype(str)
df.head(2)
```

```python
cleaned_df = df[["feels_like","weather_main","datetime"]]
cleaned_df.head()
```
